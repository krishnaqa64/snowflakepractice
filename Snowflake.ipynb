```python
usa_can=df.loc[filt2,['Country','Ethnicity','Age','Gender','EdLevel','LanguageWorkedWith',
              'LanguageDesireNextYear','YearsCode','WorkWeekHrs','ConvertedComp']]

filt3=usa_can['LanguageWorkedWith'].str.contains("JavaScript",na=False) | usa_can['LanguageWorkedWith'].str.contains("Python",na=False)

usa_can=usa_can[filt3]
usa_can
```
